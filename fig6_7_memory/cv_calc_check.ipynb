```python
# compue avg exp of mem gene in each sublg for viz

import re
import statistics

def avg_comp(in_dict, df, gene_lst):
    
    avg_dict = {}
    
    for node, str_lst in in_dict.items():
        
        #print(str_lst)
        cln = str_lst.replace("['", "").replace("']", "").replace("',", "").replace("'", "")
        
        lst = cln.split()
        
        #print(lst)
        
        clean_lst = []
        
        for bc in lst:
            
            #print(bc)
            
            if bc in list(df):
                
                clean_lst.append(bc)      
        #print(clean_lst)
        
        #print(len(lst))
        
        if clean_lst != None:
            
            #print(lst)
            sub_df = df.loc[gene_lst, clean_lst]
            col_chk = len(sub_df.columns)
            
            if col_chk > 0:
                
                sub_cells = list(sub_df)
                
                sum_lst = []
                
                for sub_cell in sub_cells:
                    
                    sries = sub_df[sub_cell]
                    sries_sum = sub_df[sub_cell].sum()
                    #print(sries_sum)
                    sum_lst.append(sries_sum)
                
                #print(sum_lst)
                med = round(statistics.median(sum_lst), 3)
                #print(med)
                
                avg_dict[node] = med
                #if sum_calc == 0: # all zero-vals
                
                #    avg_dict[node] = 0
            
                #else:
                
                #    nzero = sub_df[sub_df!=0]
                #    avg = round(nzero.median(), 3)
                #    avg_dict[node] = avg
            
    #return pd.Series(avg_dict)
    
    return pd.Series(avg_dict)
    
#egfr = avg_comp(mydict, in_1, "EGFR")

neuron_sig = avg_comp(mydict, in_1, lst)
print(neuron_sig)
```
