```python
# combine null values with experimental values in combodf 

def combo_df(df_dict, null_sries, thresh):
    
    combo_df_dict = {}
    
    for sample, df in df_dict.items():
        
        combo_df = pd.DataFrame(index = null_sries.index)
        
        combo_df["min_sub_lg_var"] = df["min_sub_lg_var"]
        combo_df["num_cells_expressing"] = df["num_cells_expressing"]
        combo_df["rand_sublg_covar_mean"] = null_sries
        print(len(combo_df))
        thresh_df = combo_df.loc[combo_df["num_cells_expressing"] >= thresh]
        print(len(thresh_df))
        combo_df_dict[sample] = thresh_df

    #print(combo_df_dict)
    return combo_df_dict

nwh_9_combo = combo_df(nwh_9_in, standard, 0)
nwh_10_combo = combo_df(nwh_10_in, standard, 0)
nwh_11_combo = combo_df(nwh_11_in, standard, 0)
nwh_12_combo = combo_df(nwh_12_in, standard, 0)
nwh_13_combo = combo_df(nwh_13_in, standard, 0)
nwh_14_combo = combo_df(nwh_14_in, standard, 0)
#nwh_14_dist_var_combo = combo_df(nwh_14_33_dist_var, standard, 0)

    
```

```python
# generate memory index dataframe

def gen_mem_index_df(df_dict, index):
    
    mem_df = pd.DataFrame(index = index, columns = list(df_dict.keys()))
    #print(mem_df)
    
    for sample, df in df_dict.items():
        
        mem_index = df["rand_sublg_covar_mean"] - df["min_sub_lg_var"]
        mem_df[sample] = mem_index
        #mem_df[sample + "_num_cells_exp"] = df["num_cells_expressing"]
    
    print(mem_df)

    return mem_df
    
nwh_9_mem_df = gen_mem_index_df(nwh_9_combo, standard.index)
nwh_10_mem_df = gen_mem_index_df(nwh_10_combo, standard.index)
nwh_11_mem_df = gen_mem_index_df(nwh_11_combo, standard.index)
nwh_12_mem_df = gen_mem_index_df(nwh_12_combo, standard.index)
nwh_13_mem_df = gen_mem_index_df(nwh_13_combo, standard.index)
nwh_14_mem_df = gen_mem_index_df(nwh_14_combo, standard.index)
#nwh_14_dist_var_mem_df = gen_mem_index_df(nwh_14_dist_var_combo, standard.index)

```
